```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_tracks=pd.read_csv("tracks.csv")
df_tracks.head()
```

```python
# Checking for null values
df_tracks.isnull().sum()
```

```python
df_tracks.info()
```

```python
df_tracks.sort_values("popularity",ascending=False)["artists"]
```

```python
#Least popular artists: 10
sorted_df=df_tracks.sort_values("popularity",ascending=True)[["artists"]].head(10)
sorted_df
```

```python
df_tracks.describe().transpose()
```

```python
#top 10 songs greater than popularity 90
filt=df_tracks["popularity"]>90
df=df_tracks.loc[filt,["name","artists","popularity"]]
most_popular=df.sort_values("popularity",ascending=False).head(10)
```

```python
#setting release_date as index and making it a datetime object
df_tracks["release_date"]=pd.to_datetime(df_tracks["release_date"],errors="coerce")
df_tracks.set_index("release_date",inplace=True)
```

```python
df_tracks.head(3)
```

```python
#converting millsec to sec and making new duration column. Removing duration_ms using drop()
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x:round(x/1000))
df_tracks.drop(columns="duration_ms",inplace=True)
```

```python
#number of songs per year
x=df_tracks.reset_index()
x["date"]=x["release_date"].dt.year
graph=sns.displot(x,x="date",binwidth=1,discrete=True,aspect=4,color="blue",kde=True).set(title="Number of Songs Per Year")

```

```python
#dropping objects for corr( ) to work
corr_df=df_tracks.drop(columns=["id","name","key","mode","explicit","artists","id_artists"])

corr_df.info()
```

```python
#Correlation heatmap
corr_df=corr_df.corr(method="pearson")
plt.figure(figsize=(10,8))
heatmap=sns.heatmap(corr_df,annot=True,cmap="rocket",fmt='.2g',linewidths=1,linecolor="black").set(title="Correlation Heatmap between numerical variables")
```

```python
plt.figure(figsize=(10,7))
sns.regplot(corr_df,y="popularity",x="acousticness",color="brown")
```

```python
#This is a common technique used to reduce the size of a dataset for various purposes such as data exploration, model training, or testing.
#loudness vs popularity
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
plt.figure(figsize=(10,7))
sns.regplot(sample_df,y="popularity",x="loudness",color="brown").set(title="Correlation between loudness and popularity")
```

```python
#Correlation between loudness and popularity
plt.figure(figsize=(10,7))
sns.regplot(sample_df,x="loudness",y="danceability",color="green").set(title="Correlation between loudness and popularity")
```
